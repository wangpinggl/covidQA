```python
import pandas as pd
import sqlite3
import numpy as np

conn = sqlite3.connect('COVIDQA.db')
c = conn.cursor()
```

```python
"""
Creates/Updates database 1 global tables from January 22,2020 to March 21,2020
"""

import datetime 
import urllib3
dateList = []
startDate = datetime.date(2020,1,22)
endDate = datetime.date(2020,3,22)
delta = endDate - startDate

for i in range(delta.days):
    day = startDate + datetime.timedelta(days=i)
    dayString = day.strftime("%m-%d-%Y")
    print(dayString)
    c.execute("SELECT name FROM sqlite_master WHERE type='table'")
    table_List = c.fetchall() 
    new_Table_List = []
    for j in range(len(table_List)):
        new_Table_List.append(table_List[j][0])
    check_String = 'db1_{}'.format(dayString.replace("-","_"))
    print(check_String)
    if check_String in new_Table_List:
        print(True)
        continue
    else:
        df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/{}.csv".format(dayString))
        df = df[['Province/State', 'Country/Region', 'Confirmed', 'Deaths', 'Recovered']]
        df = df.rename(columns = {'Province/State' : 'Province_State', 'Country/Region' : 'Country_Region'})
        query = "Create table if not exists db1_{}(Province_State text, Country_Region text, Confirmed int, Deaths int, Recovered int)".format(dayString.replace("-", "_"))
        c.execute(query)
        df.to_sql('db1_{}'.format(dayString.replace("-","_")), conn, if_exists = 'append', index = False)

c.execute("Select * from db1_01_25_2020")
print(c.fetchall())

```
