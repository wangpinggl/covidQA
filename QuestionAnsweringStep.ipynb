```python
from sumeval.metrics.rouge import RougeCalculator


rouge = RougeCalculator(stopwords=True, lang="en")

rouge_1 = rouge.rouge_n(
            summary="What county has the 2nd highest number of deaths in Virginia?",
            references="What county has the (Value Entity) number of (Case Entity) in (State Entity)?",
            n=1)

rouge_2 = rouge.rouge_n(
            summary="I went to the Mars from my living town.",
            references=["I went to Mars", "It's my living town"],
            n=2)

rouge_l = rouge.rouge_l(
            summary="I went to the Mars from my living town.",
            references=["I went to Mars", "It's my living town"])

# You need spaCy to calculate ROUGE-BE

rouge_be = rouge.rouge_be(
            summary="I went to the Mars from my living town.",
            references=["I went to Mars", "It's my living town"])

print("ROUGE-1: {}, ROUGE-2: {}, ROUGE-L: {}, ROUGE-BE: {}".format(
    rouge_1, rouge_2, rouge_l, rouge_be
).replace(", ", "\n"))
```

```python
question_dict = {}
question_entity = {}
import os
import json
import re
path = 'C:/Users/Srikar Balusu/covidQA'
real_question = {}
for file in os.listdir(path):
    if file.endswith(".json") and file != 'lookup1.json' and file != 'uniquelookup.json':
        print(file)
        with open(file) as json_file: 
            data = json.load(json_file)
        
        for i in range(len(data)):
            if len(data[str(i+1)]) == 0: 
                continue
            else:
                qt = data[str(i+1)][0]['question_template']
                query_t = data[str(i+1)][0]['query_template']
                if i+1 == 4 or i+1 == 8 or i+1 == 13 or i+1 == 25 or i+1 == 33:
                    real_q = data[str(i+1)][0]['question']
                    entity_list = data[str(i+1)][0]['entities']
                    if real_q.find("(")<0: 
                        real_question[real_q] = qt
                        question_entity[real_q] = entity_list
                if qt in question_dict.keys():
                    if query_t in question_dict[qt]:
                        continue
                    else:
                        question_dict[qt].append(query_t)
                else:
                    question_dict[qt] = []
                    question_dict[qt].append(query_t)
                    

print(question_dict)
print(real_question)
print(len(question_dict.keys()))
print(len(real_question.keys()))
print(question_dict.keys())
```
