```python
import datetime
from datetime import date, timedelta
count = 0
count2 = 0 
correct = 0
for i in range(len(data)): 
     if data[str(i+1)][0]['question_template_id'] == '2q8':
        q = data[str(i+1)][0]['question']
        ent = data[str(i+1)][0]['entities']
        if q not in tapas_data_2q8.keys():
            continue
        query = data[str(i+1)][0]['sql']
        if "African-American" in q: 
            query = query.replace("LatinX", "Black")
        if "LatinX" in query: 
            query = query.replace("LatinX", "Latinx")
        day = query.split(" ")[7].replace("'", "")
        day = int(day[-2:])
        test_date = date(2021, 1, day)
        if test_date.weekday() <= 2: 
            test_date = test_date + timedelta(days=2-test_date.weekday())
        else: 
            test_date = test_date + timedelta(days=6-test_date.weekday())
        day = query.split(" ")[7].replace("'", "")
        query = query.replace(day, test_date.strftime("%Y%m%d"))
        print(q)
        print(ent)
        print(test_date)
        print(query)
        c.execute(query)
        res = c.fetchall()
        print("Query Result:")
        print(res)
        print("Tapas Result:")
        print(tapas_data_2q8[q])
        temp = tapas_data_2q8[q].split(", ")
        print(res[0][0])
        print(temp)
        count1 = 0
        check = False
        for i in temp: 
            if count1 >= 1:
                break
            if i == str(res[0][0]): 
                check = True
            if i == 'nan' and str(res[0][0]) == 'None':
                check = True
            count1 +=1
        if check == True: 
            correct += 1
            print("Correct")
        count += 1
print(correct)
print(correct*100.0/280)
print(count)
```
